```python
import pandas as pd

customers = pd.read_csv(r'data/noahs-csv/noahs-customers.csv')
products = pd.read_csv(r'data/noahs-csv/noahs-products.csv')
orders = pd.read_csv(r'data/noahs-csv/noahs-orders.csv')
orders_items = pd.read_csv(r'data/noahs-csv/noahs-orders_items.csv')
```

```python
products[products['desc'].str.lower().str.contains("cat")]
```

```python
customers_select = customers[customers['citystatezip'].str.contains("Queens Village")]
customers_select = customers_select.merge( orders, on="customerid" )
customers_select = customers_select.merge( orders_items, on="orderid" )
customers_select = customers_select.merge(
        products[products["desc"].str.lower().str.contains("cat")],on="sku"
)
```
