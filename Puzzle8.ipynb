```python
products[products['desc'].str.contains("Noah")]
```

```python
customers_orders = customers.merge( orders, on="customerid" ).merge( orders_items, on="orderid" )
```

```python
customers_orders.merge(products[products['desc'].str.contains("Noah")], on="sku").groupby(['name']).count().sort_values('customerid', ascending=False)
```
