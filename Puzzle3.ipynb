```python
import pandas as pd

customers = pd.read_csv(r'data/noahs-csv/noahs-customers.csv')
```

```python
# Contractor from Puzzle 2
contractor = "Jeremy Davis"
print(customers[customers["name"]==contractor])
ZIP = int(list(customers[customers["name"]==contractor]['citystatezip'].str.split(" "))[0][-1])
ZIP
```

```python
# Customers born in dog years
dogyears = [ i for i in range(1910,2030,12) ]
customers['birthyear'] = pd.to_datetime(customers['birthdate']).dt.year
customers_dogyears = customers[customers['birthyear'].isin(dogyears)]
customers_dogyears
```

```python
mask = False
for y in dogyears:
    datemin = "{}-03-21".format(y) 
    datemax = "{}-04-19".format(y)
    mask = mask | ( (pd.to_datetime(customers.birthdate) >= pd.Timestamp(datemin)) & (pd.to_datetime(customers.birthdate) <  pd.Timestamp(datemax)) )

customers_dogyears_aries = customers.loc[mask]
customers_dogyears_aries
```

```python
zipquery = "zipcode == '{}'".format(ZIP)

customers_dogyears_aries_ZIP = customers_dogyears_aries.assign(
    zipcode=customers_dogyears_aries["citystatezip"].str.split(" ").str[-1]
).query(zipquery)

customers_dogyears_aries_ZIP
```
