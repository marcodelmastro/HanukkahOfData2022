```python
import pandas as pd

customers = pd.read_csv(r'data/noahs-csv/noahs-customers.csv')
products = pd.read_csv(r'data/noahs-csv/noahs-products.csv')
orders = pd.read_csv(r'data/noahs-csv/noahs-orders.csv')
orders_items = pd.read_csv(r'data/noahs-csv/noahs-orders_items.csv')
```

```python
orders
```

```python
mask1 = pd.to_datetime(orders["ordered"]).dt.hour < 5 
mask2 = pd.to_datetime(orders["shipped"]).dt.hour < 5

orders.loc[mask1 & mask2]
```

```python
# selection on order time masks
customers_select = customers.merge( orders.loc[mask1 & mask2] )

# Merge with ordered items (this will add SKU and unit_price columns)
customers_select = customers_select.merge( orders_items, on="orderid" )

# Merge with products via sku, selecting for bagel
customers_select = customers_select.merge(
        products[products["desc"].str.lower().str.contains("bagel")],on="sku"
)

customers_select
```
