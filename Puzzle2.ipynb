* **‘2017’** is the year the item was brought in, and **‘JD’** is the initials of the contractor.

```python
import pandas as pd

customers = pd.read_csv(r'data/noahs-csv/noahs-customers.csv')
products = pd.read_csv(r'data/noahs-csv/noahs-products.csv')
orders = pd.read_csv(r'data/noahs-csv/noahs-orders.csv')
orders_items = pd.read_csv(r'data/noahs-csv/noahs-orders_items.csv')
```

There's one coffee ("Coffee, Drip") and many coffee-related objects, and two bagels ("Sesame Bagel","Caraway Bagel"):

```python
products[products['desc'].str.lower().str.contains('coffee|bagel')]
```

```python
products[products['desc'].str.lower().str.contains('coffee, drip|bagel')]
```

```python
# Add 'initials' column

customers.assign(
    initials=customers["name"].str.split(" ").str[0].str[0]+customers["name"].str.split(" ").str[1].str[0]
)
```

```python
# Add 'year' column

orders = orders.assign(
    year=orders["ordered"].str.split("-").str[0]
)

orders[orders['year']=="2017"]
```

```python
# Add initiatials, select customers with JD initial

customers_select = customers.assign(
    initials=customers["name"].str.split(" ").str[0].str[0]+customers["name"].str.split(" ").str[1].str[0]
).query('initials == "JD"')

# Add year, select orders placed in 2017
customers_select = customers_select.merge( orders[orders['year']=="2017"], on="customerid" )

# Merge with ordered items (this will add SKU and unit_price columns)
customers_select = customers_select.merge( orders_items, on="orderid" )

customers_select
```

```python
# Merge with products via sku, selecting for coffee and bagel

customers_select = customers_select.merge(
        products[products["desc"].str.lower().str.contains("coffee, drip|bagel")],on="sku"
)

customers_select 
```
