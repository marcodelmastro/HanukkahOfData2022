```python
import pandas as pd

customers = pd.read_csv(r'data/noahs-csv/noahs-customers.csv')
products = pd.read_csv(r'data/noahs-csv/noahs-products.csv')
orders = pd.read_csv(r'data/noahs-csv/noahs-orders.csv')
orders_items = pd.read_csv(r'data/noahs-csv/noahs-orders_items.csv')
```

```python
womanname = "Emily Randolph"
womanID = int(customers[customers["name"]==womanname]["customerid"])
womanID
```

```python
orders_woman = orders[orders['customerid']==womanID]
orders_woman
```

```python
items_woman = orders_items.merge(orders_woman, on="orderid").merge(products,on="sku")
items_woman_colors = items_woman[items_woman['desc'].str.contains("\(")]
items_woman_colors = items_woman_colors.assign(
    day=items_woman_colors['ordered'].str.split(" ").str[0]
)
items_woman_colors
```

```python
days     = items_woman_colors['day']
ordereds = items_woman_colors['ordered']
items    = items_woman[items_woman['desc'].str.contains("\(")]['desc'].str.split(" \(").str[0] #.drop_duplicates()
colors   = items_woman[items_woman['desc'].str.contains("\(")]['desc'].str.split(" ").str[-1]
days,ordereds,items,colors
```

```python
orders_all = orders.merge(orders_items,on="orderid").merge(products,on="sku").merge(customers,on="customerid")
orders_all = orders_all.assign(
    day = orders_all['ordered'].str.split(" ").str[0]
).assign(
    objectname = orders_all['desc'].str.split(" \(").str[0]
)
#orders_all
```

```python
items    = list(items)
days     = list(days)
colors   = list(colors)
ordereds = list(ordereds)

for name,customerID,desc,day,ordered,objectname in zip(orders_all['name'],orders_all['customerid'],orders_all['desc'],orders_all['day'],orders_all['ordered'],orders_all['objectname']):
    if customerID==womanID:
        continue
    if objectname in items: # same item
        color = desc.split(" ")[-1]
        i = items.index(objectname)
        if day==days[i] and color!=colors[i]: # same day and different colots
            time_w = pd.Timestamp(ordereds[i])
            time_m = pd.Timestamp(ordered)
            dtime = pd.Timedelta("00:30:00") # bought items within half an hour
            if abs(time_m-time_w)<dtime:
                print("{} {} \t{:30s} \t{} \t{:10s} \t{}".format(customerID,name,desc,ordered,colors[i],ordereds[i]))
```
