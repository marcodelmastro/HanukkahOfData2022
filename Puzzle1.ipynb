__"if you dialed the numbers corresponding to the letters in their name, it would call their phone number!"__

- Phone numbers have 3+3+4=10 digits
- 0 and 1 don's map to any character, maybe 1 maps to space?
- Keypad:
    - ABC: 2
    - DEF: 3
    - GHI: 4
    - JKL: 5
    - MNO: 6
    - PQRS: 6
    - TUV: 8
    - WXYZ: 9

```python
from collections import defaultdict

keypad = {"ABC":  2,
          "DEF":  3,
          "GHI":  4,
          "JKL":  5,
          "MNO":  6,
          "PQRS": 7,
          "TUV":  8,
          "WXYZ": 9} 

keyValues = list(keypad.keys())

def letter2value(L):
    for K in keyValues:
        if L in K:
            return keypad[K]

letterValue = defaultdict(lambda: 0)
for L in "ABCDEFGHIJKLMNOPQRTSUVWXYZ":
    letterValue[L] = letter2value(L)
    
letterValue[" "] = 1
```

```python
def name2phone(name):
    return "".join(str(i) for i in [ letterValue[N] for N in name.upper() ] )

for name,phone,citystatezip in zip(df[df.columns[1]],df[df.columns[5]],df[df.columns[3]]):
    
    # select only NY residents
    #state,_ = citystatezip.split(", ")[1].split(" ")
    #if state != "NY": continue
    
    # remove dash to phone:
    phone_j = "".join(phone.split("-")) 
    
    # use only surname
    names = name.split(" ")
    name_ = names[1] 
    
    number = name2phone(name_)
    if str(number) == phone_j:
        print(name,number,phone)
        break
```
